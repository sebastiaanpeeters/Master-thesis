```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_fleurs = pd.read_csv('evaluation_output/model_metrics.csv')
df_fleurs
```

```python
df_nchlt = pd.read_csv('evaluation_output_nchlt/model_metrics.csv')
df_nchlt
```

```python
df_cv = pd.read_csv('evaluation_output_common_voice/model_metrics.csv')
df_cv
```

```python
full_dict = {
    'model' : df_fleurs.model,
    'model_name': ['Whisper Small', 'AF Fleurs', 'AF+DE', 'AF+NL', 'AF+NCHLT', 'AF+DE 5h', 'AF+SH', 'AF+FL', 'AF+SH v2', 'AF+FL v2'],
    'wer_fleurs': df_fleurs.wer,
    'wer_cv' : df_cv.wer,
    'wer_nchlt': df_nchlt.wer,
    'cer_fleurs': df_fleurs.cer,
    'cer_cv' : df_cv.cer,
    'cer_nchlt': df_nchlt.cer,
}
```

```python
df = pd.DataFrame(full_dict)
```

```python
df
```

```python
df_wer = df[['model_name','wer_fleurs', 'wer_cv', 'wer_nchlt']]
```

```python
df_wer = df_wer.set_index('model_name')
df_wer
```

```python
df_wer.sort_values(by='wer_fleurs')
```

```python
df_wer.wer_fleurs.sort_values()[:3]
```

```python
df_wer.wer_nchlt.sort_values()[:3]
```

```python
df_wer.wer_cv.sort_values()[:3]
```

```python
plt.figure()
lines = []  
for index, row in df_wer.iterrows():
    line, = plt.plot(['FLEURS', 'Common Voice', 'NCHLT'], row, linestyle='--', marker='o', label=index)
    lines.append(line)
    
plt.title('WER on each test set')
plt.ylabel('WER')
plt.legend()

for line in lines:
    print(f"Label: {line.get_label()}, Color: {line.get_color()}")

plt.show()
#plt.savefig('wers.png')
```

```python
df_cer = df[['model_name','cer_fleurs', 'cer_cv', 'cer_nchlt']]
df_cer = df_cer.set_index('model_name')

plt.figure()
for index, row in df_cer.iterrows():
    plt.plot(['FLEURS', 'Common Voice', 'NCHLT'], row, linestyle='--', marker='o', label=index)

plt.title('CER on each test set')
plt.ylabel('CER')
plt.legend()
#plt.show()
plt.savefig('cers.png')
```

```python
df_cer.index
```

```python
relevant_idxs = ['AF+DE', 'AF+NL', 'AF+NCHLT', 'AF+SH v2', 'AF+FL v2']
colors = {'AF+DE':'#2ca02c', 'AF+NL':'#d62728', 'AF+NCHLT':'#9467bd', 'AF+SH v2':'#bcbd22', 'AF+FL v2':'#17becf'}
```

```python
plt.figure()
for index, row in df_wer[df_wer.index.isin(relevant_idxs)].iterrows():
    plt.plot(['FLEURS', 'Common Voice', 'NCHLT'], row, linestyle='--', marker='o', label=index, color=colors[index])

plt.title('WER on each test set')
plt.ylabel('WER')
plt.legend()
plt.show()
#plt.savefig('wers_sel_v2.png')
```

```python
plt.figure()
for index, row in df_cer[df_cer.index.isin(relevant_idxs)].iterrows():
    plt.plot(['FLEURS', 'Common Voice', 'NCHLT'], row, linestyle='--', marker='o', label=index, color=colors[index])

plt.title('CER on each test set')
plt.ylabel('CER')
plt.legend()
plt.show()
#plt.savefig('cers_sel_v2.png')
```

```python
df_wer['rank_fleurs'] = df_wer['wer_fleurs'].rank(method='min', ascending=True).astype(int)
df_wer['rank_cv'] = df_wer['wer_cv'].rank(method='min', ascending=True).astype(int)
df_wer['rank_nchlt'] = df_wer['wer_nchlt'].rank(method='min', ascending=True).astype(int)
df_wer
```

```python
plt.figure()
for index, row in df_wer[['rank_fleurs', 'rank_cv', 'rank_nchlt']].iterrows():
        plt.plot(['FLEURS', 'Common Voice', 'NCHLT'], row, linestyle='--', marker='o', label=index)

plt.title('Rank among all models on each test set for WER')
plt.ylabel('Rank')
plt.legend()
plt.show()
```

```python
df_cer['rank_fleurs'] = df_cer['cer_fleurs'].rank(method='min', ascending=True).astype(int)
df_cer['rank_cv'] = df_cer['cer_cv'].rank(method='min', ascending=True).astype(int)
df_cer['rank_nchlt'] = df_cer['cer_nchlt'].rank(method='min', ascending=True).astype(int)
df_cer
```

```python
plt.figure()
for index, row in df_cer[['rank_fleurs', 'rank_cv', 'rank_nchlt']].iterrows():
        plt.plot(['FLEURS', 'Common Voice', 'NCHLT'], row, linestyle='--', marker='o', label=index)

plt.title('Rank among all models on each test set for CER')
plt.ylabel('Rank')
plt.legend()
plt.show()
```
